### 01. Importing Libraries and Data

##### 01.01. What is `Supervised Learning`?



Supervised learning is one type of machine learning where values are predicted which are already known. 

##### 01.02. Explain Supervised Learning with Metaphor.

### 02. Creating features and target arrays

```python
X = diabetes_df.drop("glucose",axis=1).values
y = diabetes_df["glucose"].values
print(type(X),type(y))
```

```python
print(y.shape,X_bmi.shape)
```

This is called **array slicing**. It means you want to select all rows (:) from the 4th column (index 3). In Python and most programming languages, indexing starts at 0, so the 4th column has an index of 3.

### 04. Reshape()

### 05. Plotting Glucose vs. BMI

```python
import matplotlib.pyplot as plt
plt.scatter(X_bmi,y)
plt.xlabel("Body Mass Index")
plt.ylabel("Body Glucose(mg/dl)")
plt.show()
```

### 06. Fit Regression Model

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X_bmi,y)
predictions = reg.predict(X_bmi)
plt.scatter(X_bmi,y)
plt.plot(X_bmi,predictions,color="black")
plt.xlabel("Body Mass Index")
plt.ylabel("Body Glucose(mg/dl)")
plt.show()
```

##### 08.01. How does `loss funtion/cost function` work?

First, we will find out the vertical distance between predictions by regression model and target values. The distance is called residual. 

Our target is, to minimize the residual as much as possible. The more we minimize the value of loss funtion, the more we will get a more accurate regression model. 

##### 08.01. What is RSS (Residual Sum of Squares)?

The sum of the squares of the difference between the target values and predictions by the regression model. 

##### 08.02. RSS and OLS (Ordinary Least Squares) are not same.


1. RSS: RSS (Residual Sum of Squares) is a measurement of prediction error. RSS is the sum of squares of the difference between predictions by the regression model and the target values. 


2. OLS: OLS (Ordinary Least Squares) is a method to find out the coefficients that minimize the RSS, thereby to identify the best-fitting linear regression model.

### 09. Linear Regression using all features

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
reg_all = LinearRegression()
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.4,random_state=42)
reg_all.fit(X_train,y_train)
y_pred = reg_all.predict(X_test)
y_pred
```

### 14. Cross-validation. Why do we need cross-validation?

The performance of a machine learning model can vary depending on how the data is split into training and test sets. Sometimes, a single random split might not provide an accurate representation of the model's generalization ability. Cross-validation is a technique that helps mitigate this issue by systematically dividing the data into multiple subsets for training and testing. It provides a more comprehensive evaluation of the model's performance, ensuring that it can generalize well to unseen data. Cross-validation is particularly useful for assessing a model's stability and reliability.

##### 14.01. Explain the `Cross-validation` with a metaphor.

### 15. `Cross-Validation` in Scikit-learn

```python
from sklearn.model_selection import cross_val_score, KFold
kf = KFold(n_splits = 6, shuffle = True, random_state = 42)
reg = LinearRegression()
cv_results = cross_val_score(reg,X,y,cv=kf)
```

```python
print(cv_results)
```

##### 15.01. Finding out mean, standard daviation and 95% confidence interval of `cv_results`

```python
import numpy as np
print(np.mean(cv_results))
```

```python
print(np.std(cv_results))
```

```python
print(np.quantile(cv_results,[0.025, 0.975]))
```
