## 01.Import Library

```python
import numpy as np
```

## 02. Gradiant Descent Funtion

```python
def gradiant_descent(x,y):
    m_curr = b_curr = 0
    iterations = 1000
    n = len(x)
    learning_rate = 0.01

    
    for i in range(iterations):
        
        
        y_predicted = m_curr*x+b_curr
        md = -(2/n)*sum(x*(y-y_predicted))
        bd = -(2/n)*sum(y-y_predicted)
    
        m_curr = m_curr - learning_rate*md
        b_curr = b_curr - learning_rate*bd
    
    
        cost = (1/n)*sum([val**2 for val in (y-y_predicted)])
    
        print("m {}, b {}, cost {}, iteration no. {}".format(m_curr,b_curr,cost, i))
    

x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])

gradiant_descent(x,y)
```

## 03. Plotting

```python
import matplotlib.pyplot as plt

%matplotlib inline
def gradiant_descent_plotting(x,y):
    m_curr = b_curr = 0
    iterations = 1000
    n = len(x)
    plt.scatter(x,y,marker='+',color='red',linewidth=5)
    
    for i in range(iterations):
        y_predicted = m_curr*x+b_curr
        plt.plot(x,y_predicted,color='green')
        
        
        md = -(2/n)*sum(x*(y-y_predicted))
        bd = -(2/n)*sum(y-y_predicted)
        
        m_curr = m_curr - 0.01 * md
        b_curr = b_curr - 0.01 * bd
        


x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])

gradiant_descent_plotting(x,y)
```
